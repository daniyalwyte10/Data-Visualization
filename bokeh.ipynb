```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show , output_file

start = datetime.datetime(2019,12,14)
end = datetime.datetime(2020,7,14)
df=data.DataReader(name="ORCL", data_source="yahoo", start=start, end=end)
df
```

```python
date_inc=df.index[df.Close > df.Open]
date_dec=df.index[df.Close < df.Open]
date_inc
```

```python
def inc_dec(c, o):
    if c > o:
        value = "Increase"
    elif c < o:
        value = "Decrease"
    else:
        value= "Equal"
    return value
df["Status"]=[inc_dec(c, o) for (c, o) in zip(df.Close, df.Open)]
df["Average"]=(df.Close+df.Open)/2
df["Height"]=abs(df.Close-df.Open)
df
```

```python
p=figure(x_axis_type='datetime', height=300, width=1000, sizing_mode="scale_width")
p.title.text='CandleStick Chart'
p.grid.grid_line_alpha=0.5

p.segment(df.index,df.High,df.index,df.Low)

hours_sec = 12*60*60*1000
p.rect(df.index[df.Status=="Increase"],df.Average[df.Status=="Increase"],
       hours_sec, df.Height[df.Status=="Increase"], fill_color="green", line_color="black")

p.rect(df.index[df.Status=="Decrease"],df.Average[df.Status=="Decrease"], 
       hours_sec, df.Height[df.Status=="Decrease"], fill_color="red", line_color="black")

output_file(r'CS.html')
show(p)
```
